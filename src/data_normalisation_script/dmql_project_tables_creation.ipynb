### amenities table

```python
from collections import defaultdict
amn_dict = defaultdict(int)
for i in df.amenities.values:
    amn_ls = i.strip("{").strip("}").split(",")
#     for i in range(len(amn_ls)):
#     amn_ls[i] = amn_ls[i].strip("\"")
    for i in amn_ls:
        amn_dict[i.strip("\"")] += 1
```

```python
amn_dict
```

```python
amn_temp_df = pd.DataFrame({'amenities':amn_dict.keys(), 'count':amn_dict.values()})
top_amenities = amn_temp_df.sort_values(by=['count'], ascending=False)['amenities'][:10].values
```

```python
top_amenities
```

```python
amn_list = []
from itertools import combinations
for i in range(1,11):
    comb = combinations(top_amenities, i)
    for j in comb:
        amn_list.append(sorted(list(j)))
```

```python
amn_list
```

```python
amn_id_df = pd.DataFrame({'amenities_id' : range(1023),'amenities_description':amn_list})
amn_id_df
```

```python
# creating a score dictionary
score_dict = {
    'Wireless Internet':10, 'Heating':9, 'Kitchen':8, 'Smoke Detector':7,
       'Essentials':6, 'Dryer':5, 'Washer':4, 'Internet':3, 'Shampoo':2, 'TV':1
}
score_dict
```

```python
# getting the descrption and score
col_amn = []
score_ls = []
for i in df.amenities.values:
    amn_ls = i.strip("{").strip("}").split(",")
#     for i in range(len(amn_ls)):
#     amn_ls[i] = amn_ls[i].strip("\"")
    value_ls = []
    score = 0
    for i in amn_ls:
        if i.strip("\"") in top_amenities:
            value_ls.append(i.strip("\""))
            score += score_dict[i.strip("\"")]
    col_amn.append(sorted(value_ls))
    score_ls.append(score)
```

```python
# updating the amenities column in the original df so that join can be performed
df['amenities'] = col_amn
```

```python
df['amenities'] = df['amenities'].astype(str)
```

```python
score_ls_2 = []
for i in amn_id_df['amenities_description'].values:
    score = 0
    for j in i:
        score += score_dict[j]
    score_ls_2.append(score)
```

```python
amn_id_df['amenities_score'] = score_ls_2
```

```python
amn_id_df['amenities_description'] = amn_id_df['amenities_description'].astype(str)
```

```python
# updating the original table
df = df.merge(amn_id_df[['amenities_id','amenities_description']], left_on='amenities', right_on=['amenities_description'], how='left')

df
```
